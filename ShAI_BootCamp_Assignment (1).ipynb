# Data Cleaning
Question 3

```python
#we see that the two columns('Status','Notes') have null values only so itis useless
df.drop(['Status','Notes'],axis=1,inplace=True)
# because these two columns('Benefits','BasePay') aren't distributed normally so we replace the null values with median
df['Benefits'].fillna(df['Benefits'].median(),inplace=True)
df['BasePay'].fillna(df['BasePay'].median(),inplace=True)
# there were another two columns('OvertimePay','OtherPay')that have 4 row of null values so we dropped it wont affect
df.dropna(inplace=True)
# we dropped the duplicated values if there
df.drop_duplicates(inplace=True)
```

# Data Visualization
Question 4

```python
df['JobTitle'].unique()
```

```python
sns.countplot(x='Year',data=df)
#2014 is highest than the others
```

```python
plt.title('The ditribution of Total Pay over the years')
sns.histplot(data=df,x='Year',y='TotalPay')
#they almost have the same range if total pay
```

```python
df2 = df[df['JobTitle'].str.contains('DEPARTMENT')]
group_words = ['POLICE DEPARTMENT', 'FIRE DEPARTMENT', 'DEPARTMENT HEAD','MEDICAL STAFF SERVICES DEPARTMENT']
df2['DEPARTMENT'] = df2['JobTitle'].apply(lambda x: next((word for word in group_words if word in x), 'Other'))
explode = (0, 0.1, 0, 0)
t=df2['DEPARTMENT'].value_counts()
plt.pie(t,startangle=180,labels=t.index,explode=explode, autopct='%1.1f%%')

# we can see that the police department is the largest one and the smallest one is the medical department
```

```python
sns.barplot(data=df2,y='TotalPay',x='DEPARTMENT')
plt.xticks(rotation=45)

# head department is the the highest in total pay
```

```python
plt.title('The ditribution of Total Pay Benefits and OvertimePay')
sns.histplot(data=df,y='OvertimePay',x='TotalPayBenefits')
# there is direct relation between the over time and total pay benefits
```

```python
plt.title('The ditribution of Total Pay and Total Pay benefits')
sns.histplot(data=df,x='TotalPay',y='TotalPayBenefits')
# there is direct relation between Total Pay and Total Pay benefits
```
